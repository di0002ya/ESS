## 2. Analysis Relationship Between Each Features

### 2.1 Temperature vs Capacity

Let's look at the coorelation coefficient. 

```python
import numpy as np 
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def generate_plot(nameX,nameY,df,df_name):
    x = np.array(df[[nameX]])
    Y = np.array(df[[nameY]])
    %matplotlib inline
    matplotlib.style.use('ggplot')
    plt.scatter(x, Y)
    plt.title(df_name+':'+nameX+'-'+nameY)
    plt.show()
```

```python
generate_plot('Capacity','temp_mean',train05,'train05')
generate_plot('Capacity','temp_mean',train06,'train06')
generate_plot('Capacity','temp_mean',test07,'test07')
generate_plot('Capacity','temp_mean',test18,'test18')
```

### 2.2 Chargecycle vs Capacity

```python
generate_plot('Capacity','charge_cycle',train05,'train05')
generate_plot('Capacity','charge_cycle',train06,'train06')
generate_plot('Capacity','charge_cycle',test07,'test07')
generate_plot('Capacity','charge_cycle',test18,'test18')
```

### 2.3 Delta vs Capacity

```python
def generate_delta_single(i,df,df_name):
    x = np.arange(35)
    Y = df.iloc[i][1:36]
    capacity = df.iloc[i][0]
    %matplotlib inline
    matplotlib.style.use('ggplot')
    plt.scatter(x, Y)
    plt.title(df_name+'-->'+'Capacity:'+str(capacity))
    plt.show()
```

```python
generate_delta_single(0,train05,'train05')
generate_delta_single(50,train05,'train05')
generate_delta_single(100,train05,'train05')
```
